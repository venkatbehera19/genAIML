```python
from pydantic import BaseModel

class User(BaseModel):
  name: str
  age: int
  email: str

user = User(name='Venkat', age=20, email='v.behera')

print(user)
```

```python
user = User(name=20)
```

```python
### nested

class Address(BaseModel):
  city: str
  pincode: int

class User(BaseModel):
  name: str
  address: Address

data = {
  "name": 'Behera',
  "address": {
    "city": 'bbsr',
    "pincode": 768110
  }
}

user = User(**data)
```
