```python
### it allows a class to inherit the attributes and methods from another class

class Animal:
  def __init__(self, name):
    self.name = name

  def info(self):
    print('Animal Name is ', self.name)

class Dog(Animal):
  def __init__(self, name, bread):
    super().__init__(name)
    self.bread = bread

  def sound(self):
    print(self.name, 'barks')

  def details(self):
    print(self.name, " is a ", self.bread)

# dog = Dog("Buddy")
# dog.sound()
# dog.info()
```

```python
### super is used to call the parent class methods. 

dog2 = Dog("buddy", "golden retriver")
# dog2.info()
```

```python
dog2.sound()
dog2.info()
dog2.details()
```

```python
### single inheritance

class Person:
  def __init__(self, name):
    self.name = name

class Employee(Person):
  def show_role(self):
    print(self.name, "is an employee")

emp = Employee('Venkat')
emp.show_role()
```

```python
### multiple inheritance
class Person:
  def __init__(self, name):
    self.name = name

class Salary:
  def __init__(self, salary):
    self.salary = salary

class Employee(Person, Salary):
  def __init__(self, name, salary):
    Person.__init__(self, name)
    Salary.__init__(self, salary)

  def details(self):
    print(self.name, 'earns' , self.salary)

emp = Employee('Venkat raman', 90000)
emp.details()
```

```python
### multilevel inheritance

class Person:
  def __init__(self, name):
    self.name = name

class Employee(Person):
  def show_details(self):
    print(self.name , " is an employeee")

class Manager(Employee):
  def department(self, dept):
    print(self.name, ' manages', dept, ' department')


mang = Manager('Subham')
mang.show_details()
mang.department('HR')
```

```python
# Hybrid inheritance

class Person:
  def __init__(self, name):
    self.name = name

class Employee(Person):
  def show_details(self):
    print(self.name , " is an employeee")

class Project:
  def __init__(self, project_name):
    self.project_name = project_name


class TeamLead(Employee, Project):
  def __init__(self, name, project_name):
    Employee.__init__(self, name)
    Project.__init__(self, project_name)

  def details(self):
    print(self.name, "leads project:", self.project_name)

tl = TeamLead('Venkat', 'Somamedia')
tl.show_details()
tl.details()
```
