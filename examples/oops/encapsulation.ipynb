```python
### binding data members and method into a single unit 


class Employee:
  def __init__(self, name, salary):
    self.name = name #public
    self.__salary = salary #private

  def display_name(self):
    print('Your Name is ', self.name)

emp = Employee('Raka', 70000)
emp.name
```

```python
emp.__salary
```

```python
emp.display_name()
```

```python
### protected members can be accsable in child class
class Employee:
  def __init__(self, name, age):
    self.name = name
    self._age = age

class SubEmployee(Employee):
  def show_age(self):
    print('Your Age is ', self._age)
```

```python
se = SubEmployee('Ven', 20)
se.show_age()
```

```python
class BankAccount:
  def __init__(self):
    self.balance = 1000

  def _show_balance(self): #protected method
    print(f"Balance: ${self.balance}")

  def __update_balance(self, amount): #private
    self.balance += amount

  def deposit(self, amount): #public
    if amount > 0 :
      self.__update_balance(amount)
      self._show_balance()
    else:
      print('Invalid Amount')


ba = BankAccount()
ba.balance
```

```python
ba.deposit(200)
```

```python
ba._show_balance()
```
