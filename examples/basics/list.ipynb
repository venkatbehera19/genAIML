```python
cars = ['Mitsubishi', "FORD", 'BMW', 'KIA', 'Volvo']

cars.sort(reverse=True)
```

```python
print(cars)
```

```python
def myFunc(e):
  return len(e)

cars.sort(key=myFunc)

print(cars)
```

```python
def myFunc(e):
  return e['year']
```

```python
cars_1 = [
  {'car': 'Ford', 'year': 2005},
  {'car': 'Mitsubishi', 'year': 2000},
  {'car': 'BMW', 'year': 2019},
  {'car': 'VW', 'year': 2011}
]

cars_1.sort(key=myFunc)
```

```python
### sort a tuple
d = [ (1, 3), (2, 2), (3, 1) ]

d.sort(key=lambda x: x[1])

print(d)
```
