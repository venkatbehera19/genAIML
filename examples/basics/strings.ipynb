```python
# Question - string duplicate removal
""" 
Input: s = "geEksforGEeks"
Output: "geEksforG"
"""

def strDuplicateRemoval(input_string: str):
  res_dict = {}
  res = ""
  for char in input_string:
    if res_dict.get(char) == 1:
      continue
    else:
      res_dict[char] = 1
      res = res + char
    print(res)
    print(res_dict)

  return res

strDuplicateRemoval('Hello')
```

```python
### reverse a string
def revere(input_string: str):
  return input_string[::-1]

##without slicing
def wit_reverse(input_string: str):
  res = ""
  for i in input_string:
    res = i + res
  return res
print(revere('Hello'))
```
