```python
import asyncio

async def fn1():
  print('First Function')
  await asyncio.sleep(1)
  print('Async function done')

async def fn2():
  print('second Function')
  await asyncio.sleep(1)
  print('second function done')

async def main():
  await asyncio.gather(fn1(), fn2())

await main()
```
